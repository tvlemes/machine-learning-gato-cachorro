```python
# Bibliotecas
import pandas as pd
from sklearn import tree
%pylab inline
```

```python
# Importando os dados
df = pd.read_excel('/content/drive/MyDrive/Pos-Graduacao/datasets/classificacao_cachorro_gato.xlsx', sheet_name=1, nrows=10)
df
```

```python
x = df.drop('eh_um_gato', axis=1)
```

```python
y = df['eh_um_gato']
```

```python
dt = tree.DecisionTreeClassifier()
```

```python
# Treino
dt.fit(x,y)
```

```python
# Resposta - target
df['resposta'] = dt.predict(x)
df
```

```python
# Retorno em probabilidade - target
df['probabilidade'] = dt.predict_proba(x)[:,1]
df
```

```python
pylab.figure(figsize=(15,9))
tree.plot_tree(dt, feature_names=x.columns);
```
